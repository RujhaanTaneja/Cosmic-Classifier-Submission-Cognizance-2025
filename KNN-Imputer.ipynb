```python
total_data=pd.read_csv("/kaggle/input/cosmicdata-initial/cosmic classifier data/cosmicclassifierTraining.csv")
```

```python
def convert_category_columns(df, column_names):
    df_copy = df.copy()
    for column in column_names:
        df_copy[column] = df_copy[column].apply(lambda x: int(re.search(r'\d+', str(x)).group()) if pd.notna(x) and re.search(r'\d+', str(x)) else x)
    return df_copy
```

```python
total_data.dropna(subset='Prediction',inplace=True)
total_data=convert_category_columns(total_data,column_names=['Magnetic Field Strength','Radiation Levels'])
```

```python
x=total_data.drop('Prediction',axis=1)
y=total_data['Prediction']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.25, random_state=42)
```

```python
imputer=KNNImputer(n_neighbors=1)
```

```python
x_train=imputer.fit_transform(x_train)
```

```python
x_train=pd.DataFrame(x_train,columns=x.columns)
```
